**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import pandas as pd
import numpy as np
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
print(df)
````

````python
import seaborn as sns
plt.figure(figsize=(12, 10))
sns.boxplot(data=df2)
plt.xticks(rotation=90)
plt.title('Box Plots of Numeric Variables')
plt.show()
````

* most of the numeric variables have outliers
* so that the dataset need to be cleaned

````python
plt.scatter(df2['BMI'], df2['BloodPressure'])
plt.xlabel('BMI')
plt.ylabel('Blood Pressure')
plt.title('BMI vs Blood Pressure')
plt.show()
````

* Most data points are clustered in the BMI range of 20 to 40 and blood pressure range of 60 to 100.
* There are some outliers, particularly individuals with very low blood pressure or very high BMI.
* The plot doesn’t show a clear linear relationship between BMI and blood pressure, indicating that BMI alone might not be a strong predictor of blood pressure.

````python
corr = df2.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
````



*   **Positive correlation**


1. Pregnancies and Age (0.55)
2. Glucose and Outcome (0.48)
3. SkinThickness and Insulin (0.43)
4. BMI and SkinThickness (0.38)


*   **Negative correlation**



1. BMI and Pregnancies (-0.0057)
2. BloodPressure and DiabetesPedigreeFunction (0.026)


The matrix suggests that certain variables like Glucose, BMI, Age, and Insulin have notable correlations with the Outcome (diabetes status), which could be important factors in predicting diabetes in the dataset.

````python
import seaborn as sns

sns.boxplot(data=df2[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']])
plt.title('Distribution of Features')
plt.show()
````

````python
df2.plot.scatter(x='Glucose', y='BMI')
plt.title('Scatter Plot of Glucose vs. BMI')
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.show()
````

````python
sns.pairplot(df2, hue='Outcome')
plt.show()
````

* Glucose vs. Outcome: The plot for Glucose shows that individuals with higher glucose levels (right side of the distribution) are more likely to have Outcome=1, indicating diabetes.
* BMI vs. Outcome: Similarly, higher BMI values seem to be associated with Outcome=1.
* Insulin vs. SkinThickness: There’s a positive relationship between Insulin and SkinThickness, particularly noticeable for Outcome=1.
* Pregnancies vs. Outcome: Higher numbers of pregnancies seem to be somewhat correlated with Outcome=1.

````python
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df2['Glucose'], df2['BMI'], df2['Age'], c=df2['Outcome'], cmap='viridis')
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
ax.set_title('3D Scatter Plot of Glucose, BMI, and Age')

plt.show()
````

from this dataset we can say that the features 'Glucose', 'BMI','Pregnancies' have a positive correlation with the outcome . ie, the higher these values , tyhere is high chance that the patient is being diabetic
